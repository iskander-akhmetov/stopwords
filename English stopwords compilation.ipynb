```python
from os import listdir
import os.path 
```

```python
import pickle
def save_obj(obj, name):
    pickle.dump(obj,open(name + '.pkl', 'wb'), protocol=4)
    
def load_obj(name):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
folder = 'stopwords_compl'
files = listdir(folder)
```

```python
stopwords_list = []
for f in files:
    if os.path.isfile(folder+'/'+f):
        file = open(folder+'/'+f, 'r')
        words = file.readlines()
        file.close()
        words = [w.replace('\n','').strip() for w in words]
        stopwords_list.append(words)
```

```python
from itertools import chain
stopwords_list = list(chain(*stopwords_list))
stopwords_list
```
